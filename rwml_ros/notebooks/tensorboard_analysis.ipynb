```python
vals= ["DQN_26","DQN_27","DQN_28","DQN_31","DQN_32"]
```

```python
import requests
```

```python
import matplotlib.pyplot as plt
```

```python
results = []
for v in vals:
    addr = "http://localhost:6006/data/plugin/scalars/scalars?tag=rollout/ep_rew_mean&run=%s&format=json"%v
    resp = requests.get(url=addr)
    data = resp.json()
    results.append(data)
```

```python
for i in range(5):
    res = results[i]
    xs  = [i[1] for i in res]
    ys = [i[2] for i in res]
    plt.plot(xs,ys)
plt.xticks(ticks=[1e3,2e3,3e3,4e3,5e3],labels=["1k","2k","3k","4k","5k"])
plt.title("Episode mean reward on Robot")
plt.xlabel("Episode")
plt.ylabel("mean reward")
plt.savefig("episode_reward_robot.png")
```
